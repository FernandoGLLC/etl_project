Business Question: Which films attracted the most audience per screening, based on average occupancy?


Takeaway: Identifies high-performing films that consistently filled a larger portion of theater capacity, helping inform future film acquisition, marketing, and scheduling decisions. High occupancy with fewer total tickets may still indicate strong demand relative to availability.

MODEL 6 QUERY:

```python
query = """
SELECT 
   film_code,
   AVG(occu_perc) AS avg_occupancy_per_film,
   SUM(tickets_sold) AS tot_tickets_sold
FROM master_model
GROUP BY film_code
ORDER BY avg_occupancy_per_film DESC
"""
conn.execute(query).fetchdf()
```
