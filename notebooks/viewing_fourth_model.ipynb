MODEL 4:

What it does?

Shows monthly revenue for each cinema from feb-nov 2018 (the data available)

Business Question:

How well are cinemas performing over time, and are there seasonal trends in cinema revenue?

Takeaway: 

Helps identify whether revenue drops during certain months or seasons and which cinemas are stable in performance.



```python
import duckdb
```

```python
conn = duckdb.connect(r'..\cinema_tickets.duckdb')
conn
```

MODEL 4 QUERY:

```python
query = """
SELECT
    CAST(cinema_code AS TEXT) AS cinema_code,
    month,
    SUM(total_revenue_usd) AS monthly_revenue
FROM master_model
GROUP BY cinema_code, month
ORDER BY month
"""

conn.execute(query).fetch_df()
```

```python
fourth_model = """
SELECT *
FROM fourth_model
"""

conn.execute(fourth_model).fetchdf()
```
